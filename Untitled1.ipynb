```python
image=cv2.imread(r"C:\\Users\Student\Desktop\pic02\x.jpg")
newimage2=np.zeros(image.shape,image.dtype)
alpha=float(input('enter a alpha'))
beta=int(input('enter a beta'))
for y in range(image.shape[0]):
    for x in range(image.shape[1]):
        for c in range(image.shape[2]):
            newimage2[y,x,c]=np.clip(alpha*image[y,x,c]+beta,0,255)
plt.figure(figsize=[15,5])
plt.subplot(121);plt.imshow(image[...,::-1]);plt.title("image");
plt.subplot(122);plt.imshow(newimage2[...,::-1]);plt.title("new");
plt.show()
```

```python
gamma=0.4
img=cv2.imread(r"C:\\Users\Student\Desktop\pic02\000.jpg")
lookuptable=np.empty((1,256),np.uint8)
for i in range(256):
    lookuptable[0,i]=np.clip(pow(i/255,gamma)*255,0,255)
output=cv2.LUT(img,lookuptable)
plt.figure(figsize=[15,5])
plt.subplot(121);plt.imshow(img[...,::-1]);plt.title("Orginal");
plt.subplot(122);plt.imshow(output[...,::-1]);plt.title("Output");
plt.show()
```

```python
gamma=0.5
img01=cv2.imread(r"C:\\Users\Student\Desktop\pic02\a.jpg")
img02=cv2.imread(r"C:\\Users\Student\Desktop\pic02\aaa.jpg")

output1=cv2.convertScaleAbs(img01,alpha=1.3,beta=40)
output2=cv2.convertScaleAbs(img02,alpha=1.3,beta=40)

lookuptable=np.empty((1,256),np.uint8)
for i in range(256):
    lookuptable[0,i]=np.clip(pow(i/255,gamma)*255,0,255)
    
output1_gamma=cv2.LUT(img01,lookuptable)
output2_gamma=cv2.LUT(img02,lookuptable)

plt.figure(figsize=[15,5])
plt.subplot(231);plt.imshow(img01[...,::-1]);plt.title("Orginal");
plt.subplot(232);plt.imshow(output1[...,::-1]);plt.title("Output1");
plt.subplot(233);plt.imshow(output1_gamma[...,::-1]);plt.title("output1_gamma");
plt.subplot(234);plt.imshow(img02[...,::-1]);plt.title("Orginal");
plt.subplot(235);plt.imshow(output2[...,::-1]);plt.title("Output2");
plt.subplot(236);plt.imshow(output2_gamma[...,::-1]);plt.title("output2_gamma");

plt.show()
```

```python
import numpy as np
```

```python
square=np.zeros((10,10),np.uint8)
cv2.rectangle(square,(2,2),(6,6),2,4)
print("Square")
print(square)
```

```python
circle=np.zeros((10,10),np.uint8)
cv2.circle(circle,(5,5),4,2,-1)
print("circle")
print(circle)
```

```python
output=cv2.bitwise_and(square,circle)
print(output)
```

```python
output=cv2.bitwise_or(square,circle)
print(output)
```

```python
square1=np.zeros((300,300),np.uint8)
cv2.rectangle(square1,(50,50),(250,250),255,-2)
ellipse=np.zeros((300,300),np.uint8)
cv2.ellipse(ellipse,(150,150),(150,150),30,0,180,255,-1)
plt.subplot(121);plt.imshow(square1,cmap="gray");plt.title("Square1")
plt.subplot(122);plt.imshow(ellipse,cmap="gray");plt.title("ellipse")
plt.show()
```

```python
bitwiseand=cv2.bitwise_and(square1,ellipse)
bitwiseor=cv2.bitwise_or(square1,ellipse)
bitwisexor=cv2.bitwise_xor(square1,ellipse)
bitwisenot=cv2.bitwise_not(square1)
print("bitwiseAnd",bitwiseand)
print("bitwiseOr",bitwiseor)
print("bitwiseXor",bitwisexor)
print("bitwiseNot",bitwisenot)
```

```python
plt.subplot(141);plt.imshow(bitwiseand,cmap="gray");plt.title("bitwiseand");
plt.subplot(142);plt.imshow(bitwiseor,cmap="gray");plt.title("bitwiseor");
plt.subplot(143);plt.imshow(bitwisexor,cmap="gray");plt.title("bitwisexor");
plt.subplot(144);plt.imshow(bitwisenot,cmap="gray");plt.title("bitwisenot");
plt.show()
```
